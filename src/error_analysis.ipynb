```python
import pandas as pd
import numpy as np
```

```python
test_pred = np.loadtxt('../data/MAMI_processed/pred/0.3_ours/test.txt')
```

```python
test_pred
```

```python
test_df = pd.read_csv('../data/MAMI_processed/test.csv')
test_df
```

```python
test_df.columns
```

```python
test_df[['shaming_probs','stereotype_probs', 'objectification_probs', 'violence_probs']] = test_pred
test_df
```

```python
test_df[['shaming_cls','stereotype_cls', 'objectification_cls', 'violence_cls']] = test_df[['shaming_probs','stereotype_probs', 'objectification_probs', 'violence_probs']] > 0.5
test_df
```

```python
num_true_pred = (test_df[['shaming_cls','stereotype_cls', 'objectification_cls', 'violence_cls']].values == test_df[['shaming','stereotype', 'objectification', 'violence']].values).sum(axis=1)
```

```python
demo_df = pd.DataFrame(columns=test_df.columns)
demo_df
```

```python
# test_df[(num_true_pred==4)&(test_df['misogynous']==1)]
```

```python
# (num_true_pred == 0).sum()
```

```python
sample_df = test_df[(num_true_pred==4)&(test_df['misogynous']==1)].sample(3, random_state=42)

demo_df = pd.concat([demo_df, sample_df])
demo_df
```

```python
sample_df = test_df[(num_true_pred==4)&(test_df['misogynous']==0)].sample(2, random_state=0)

demo_df = pd.concat([demo_df, sample_df])

demo_df
```

```python
num_true_positive_pred = (test_df[['shaming_cls','stereotype_cls', 'objectification_cls', 'violence_cls']].values & test_df[['shaming','stereotype', 'objectification', 'violence']].values).sum(axis=1)
num_true_positive_pred
```

```python
num_positive = test_df[['shaming','stereotype', 'objectification', 'violence']].values.sum(axis=1)
num_positive
```

```python
sample_df = test_df[(num_true_positive_pred==2) & (num_positive==3)].sample(2, random_state=1)

demo_df = pd.concat([demo_df, sample_df])

demo_df
```

```python
demo_df.to_csv('../data/MAMI_processed/demo_df.csv', index=False)
```
