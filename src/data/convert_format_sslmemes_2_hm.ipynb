```python
import numpy as np
import pandas as pd
```

```python
def convert(root_path, split):
    df = pd.read_csv(f'{root_path}/{split}.csv')
    df['img'] = df['img'].map(lambda x: 'img/'+x)
    df.to_json(f'{root_path}/{split}.jsonl', orient='records', lines=True)
    
    print(f'Successfully converted {split} at {root_path}')
```

```python
ROOT_PATH = '../../data/hateful_memes/'

splits = ['train_labeled_ratio-0.3', 'train_unlabeled_ratio-0.3',
         'train_labeled_ratio-0.1', 'train_unlabeled_ratio-0.1',
         'train_labeled_ratio-0.05', 'train_unlabeled_ratio-0.05',]
```

```python
for split in splits:
    convert(ROOT_PATH, split)
```

```python
for split in splits:
    print(split)
    df = pd.read_json(f'{ROOT_PATH}/{split}.jsonl', lines=True)
    display(df)
```
